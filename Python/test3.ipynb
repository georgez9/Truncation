```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
map_white_matters = pd.read_csv(r".\data\connection_map_white_matter.csv", header=None).to_numpy(dtype=int)
map_grey_matters = pd.read_csv(r".\data\connection_map_grey_matter.csv", header=None).to_numpy(dtype=int)
```

```python
sns.heatmap(map_grey_matters,
            xticklabels=[],
            yticklabels=[],
            cbar=None,
            square="equal",
            cmap= ["#2e5d87", "#f7cb67"]
            )
plt.savefig('pictures/grey_matter.svg', dpi=300)
```

```python
sns.heatmap(map_white_matters,
            xticklabels=[],
            yticklabels=[],
            cbar=None,
            square="equal",
            cmap= ["#2e5d87", "#f7cb67"]
            )
plt.savefig('pictures/white_matter.svg', dpi=300)
```
