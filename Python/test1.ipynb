```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
import numpy as np
```

```python
df = pd.read_csv('mean.csv')
df.head(5)
```

```python
plt.figure(figsize=(6, 6))

sns.violinplot(data=df, x="Regions", y="Connectivity", hue="Type", palette=["#F17C67", "#90B44B"], alpha=0.6, 
               split=True, inner="quart", linewidth=1, density_norm="count")

ax = sns.stripplot(data=df, x="Regions", y="Connectivity", hue="Type",
              dodge=True, alpha=0.9, jitter=True, palette=["#880000", "#008800"], size=2)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], title='Regions')

plt.title('Distrubution of all matters')
plt.show()
```

```python
data1 = df[(df['Type'] == 'actual') & (df['Regions'] == 'after tp')]['Connectivity']
data2 = df[(df['Type'] == 'actual') & (df['Regions'] == 'across_tp')]['Connectivity']
data1 = data1.dropna()
data2 = data2.dropna()
data1.size
```

```python
statistic, p_value = ks_2samp(data1, data2)

print(f"K-S: {statistic}")
print(f"p: {p_value}")
```

```python
# data1data2KDE
sns.kdeplot(data1, color="blue", label="Before TP")
sns.kdeplot(data2, color="red", label="After TP")

plt.legend(title="Regions")
plt.xlabel("Connectivity")
plt.ylabel("Density")
plt.title("KDE of Connectivity Before and After TP")
plt.show()
```
