```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('mean.csv')
df.head(5)
```

```python
plt.figure(figsize=(6, 6))

sns.violinplot(data=df, x="Type", y="Connectivity", hue="Regions", palette=["#F17C67", "#90B44B"], alpha=0.7, 
               split=True, inner="quart", linewidth=1)

ax = sns.stripplot(data=df, x="Type", y="Connectivity", hue="Regions",
              dodge=True, alpha=0.8, jitter=True, palette=["#880000", "#008800"], size=3, linewidth=0.5)

# plt.ylim(0, 1)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], title='Regions')

plt.title('distrubution')
plt.show()
```
